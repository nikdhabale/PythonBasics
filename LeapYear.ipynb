```python
def new_func(year):
    # Get Year in 4 digits as input from the user
    #year = input ("Enter a year in YYYY format: ")
# Check if the year is a leap year or not  
    if (int(year) % 4) == 0:
        if (int(year) % 100) == 0:
            if (int(year) % 400) == 0:
                print("{0} is a leap year".format(year))
            else:
                print("{0} is not a leap year".format(year))
        else:
            print("{0} is a leap year".format(year))
    else:
        print("{0} is not a leap year".format(year))

new_func(2008)
```
