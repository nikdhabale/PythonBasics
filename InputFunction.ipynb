Input Fucntion : Essential feature in Python that allows to take input from
the user

It reads the input as a string and returns it

```python
name = input("Enter your name: ")
print("Hello, " + name + "!")
```

```python
# add two given numbers
num1 = input("Enter first number: ")
num2 = input("Enter second number: ")
sum = float(num1) + float(num2)
print("The sum of {0} and {1} is {2}".format(num1, num2, sum))
```

```python
# Student Marks Calculation
student_name = input("Enter student name: ")
marks_subject1 = float(input("Enter marks for Subject 1: "))
marks_subject2 = float(input("Enter marks for Subject 2: "))
marks_subject3 = float(input("Enter marks for Subject 3: "))
total_marks = marks_subject1 + marks_subject2 + marks_subject3
average_marks = total_marks / 3
print("Student Name: " + student_name)
print("Total Marks: " + str(total_marks))
print("Average Marks: " + str(average_marks))
```
