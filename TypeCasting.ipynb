Type Casting :
1. The process of converting a value from one data
type to another


Python has several built-in functions for type casting:
int(): Converts a value to an integer.
float(): Converts a value to a floating-point number.
str(): Converts a value to a string.
list(), tuple(), set(), dict() and bool()

```python
str = "14"
int_str = int(str)
print(int_str + 6)
print(type(int_str))
```

```python
Var = "3.14"
float_var = float(Var)
print(float_var)
print(type(float_var))
```

```python
# Converting Float to Integer:
float_num = 3.14
int_num = int(float_num)
print(int_num) 
print(type(int_num)) 
```

Types of Typecasting
There are two types of type casting in python:
• Implicit type casting : performed automatically by the Python interpreter
• Explicit type casting : Also known as type conversion, is performed manually by the programmer
using built-in functions

```python
# Implicit type casting example
int_var = 10
float_var = 5.5
result = int_var + float_var
print(result)
print(type(result))
```

```python
# explicit type casting example
float_num = 9.8
int_num = int(float_num)
print(int_num)
print(type(int_num))
```
